```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
train_data=pd.read_csv("/kaggle/input/titanic/train.csv")
test_data=pd.read_csv("/kaggle/input/titanic/test.csv")
```

```python
#feature for training
features=['Pclass','Sex','SibSp','Parch']

#turns class features into binary
x_train=pd.get_dummies(train_data[features],dtype=int)
y_train=train_data.Survived
x_test=pd.get_dummies(test_data[features],dtype=int)
```

```python
model=RandomForestClassifier(max_depth=5,random_state=1)
model.fit(x_train,y_train)
```

```python
predict_df = pd.DataFrame(model.predict(x_test),columns=['Survived'])

predict_df=predict_df.rename(index=lambda a : a+892)
predict_df.to_csv('/kaggle/working/predict2.csv')
```
